```python
# data agregatition by group by
import pandas as pd
data = {'name':['Jen','Max','Tina','John','Kay','Ish','Ami'], 'year':[2012, 2012, 2013, 2012, 2013, 2014, 2014], 'score' :[4, 5, 9, 8, 9, 5, 4]}
df = pd.DataFrame(data)
df.groupby(['year'])['score'].mean()
```

```python
df.groupby(['year'])['score'].median()
```

```python
# pandas dapat digunakan untuk statistika analisis
import pandas as pd
data = {'name':['Jason','Molly','Tina','Jake','Amy'], 'age':[42, 52, 36, 42, 73], 'score':[4, 24, 31, 2, 3]}
df = pd.DataFrame(data)
df.describe().T
```

```python
print("nilai mean");print(df['score'].mean())
print("nilai median");print(df['score'].median())
print("nilai mode");print(df['age'].mode())
print("nilai var");print(df['age'].var())
print("nilai skew");print(df['age'].skew())
```

```python
# GRAFIK PIECHART (diagram lingkaran)
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8)) # setting ukuran grafik yang diinginkan
labels = 'frogs', 'Hogs', 'Dogs', 'Logs'
fracs = [15, 30, 45, 10]
plt.pie (fracs, labels=labels, autopct='%1.1f%%', shadow=True)
plt.show() # menampilkan grafik yang sudah dibuat
```

```python
# MEMBUAT DIAGRAM BATANG

from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [0, 1, 2, 3, 4, 5]
y = [24.27, 23.18, 22.39, 8.41, 7.19, 6.62]

fig, ax = plt.subplots()

ax.bar(x, y, align='center')

ax.set_title('Bahasa Pemrograman Favorit (UNPAM)')
ax.set_ylabel('Pengguna')
ax.set_xlabel('Bahasa')

ax.set_xticks(x)
ax.set_xticklabels(("Python", "JavaScript", "Java", "C#", "PHP", "C++"))

plt.show()
```

Keterangan

style.use → Gaya grafik yang digunakan

fig,ax → Menambahkan kolom figure dan axes pada sublplots()

ax.bar → Membuat output persegi

ax.set_title → Judul grafik

ax.set_ylabel → Menamai garis vertikal

ax.set_xlabel → Menamai garis horizontal

ax.set_xticks(x) → Memberikan nilai pada garis horizontal

ax.set_xticklabels → Menamai nilai yang diciptakan xticks

plt.show → Menampilkan chart

```python
import matplotlib.pyplot as plt

y = [2990, 2710, 2540, 3300, 2990]
x = [18, 19, 20, 21, 22]
plt.plot(x, y)

plt.xlabel('Tanggal (Januari)')
plt.ylabel('Harga')

plt.title('Harga Emiten ANTM')
plt.grid(True)

plt.show()
```

Keterangan:

plt.plot(x,y) → Meletakkan posisi titik di x dan y

plt.xlabel → Keterangan sumbu horizontal

plt.ylabel → Keterangan sumbu vertikal

plt.title → Bagan judul

plt.grid → Menampilkan garis grid

plt.show → Menampilkan chart
